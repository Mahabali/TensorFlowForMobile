```python
# Import required modules
import tensorflow as tf
import numpy as np
from tensorflow import keras
from tensorflow import lite
```

```python
# Sample data
x = np.array([1.0,2.0,4.0,5.0,3.0,7.0,6.0,8.0,9.0,9.5,9.9,10.2,12.0,15.0,17.0],dtype= float)
y = np.array([3.0,5.0,9.0,11.0,7.0,15.0,13.0,17.0,19.0,20.0,20.8,21.4,25.0,31.0,35.0],dtype= float)
#Training
model = keras.Sequential([keras.layers.Dense(units=1,input_shape=[1]),keras.layers.Dense(units=1,input_shape=[1])])
model.compile(optimizer='Nadam', loss='mean_absolute_error')
model.fit(x,y,epochs=1500)
#Predicting data
print(model.predict([41.5]))
```

```python
# Save Model for future training
tensor_flow_file_name = "SampleModel.h5"
tf.keras.models.save_model(model,tensor_flow_file_name)
```

```python
# Convert to tflite use
tensor_flow_lite_file_name = "SampleTFLiteModel.tflite"
converter = lite.TFLiteConverter.from_keras_model(model)
tfmodel = converter.convert()
open(tensor_flow_lite_file_name,"wb").write(tfmodel)
```

```python
# Convert to CoreML for iOS
import coremltools as ct
core_ml_file_name = "SampleMLModel.mlmodel"
mlmodel = ct.convert(model)
spec = mlmodel.get_spec()

## Rename input and output, so that its easy to reference in coreml
# get input names
input_names = [inp.name for inp in spec.description.input]
# get output names
output_names = [out.name for out in spec.description.output]

print(input_names)
print(output_names)
# Input and Output are names for Input and Output. 
# By default input is some dense_layer_xxx and output is Identity in 
# CoreML
ct.utils.rename_feature(spec, input_names[0], 'Input')
ct.utils.rename_feature(spec, output_names[0], 'Output')


# reload the model with the updated spec and re-save
mlmodel = ct.models.MLModel(spec)
mlmodel.save(core_ml_file_name)
```
